```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import csv
import seaborn as sns
```

```python
data = os.path.join("Downloads", "yelp_academic_dataset_business.csv")
business = pd.read_csv(data)
business.head(10)
```

```python
business['categories'] = business['categories'].astype(str)
business1=business[business['categories'].str.contains('Restaurants')]
business1
```

```python
business_basics=business1[['attributes.Delivery','attributes.Dogs Allowed','attributes.Drive-Thru','attributes.Happy Hour','attributes.Noise Level','attributes.Outdoor Seating','attributes.Price Range','attributes.Takes Reservations','attributes.Wi-Fi','business_id','categories','city','full_address','latitude','longitude','name','neighborhoods','review_count','stars','state','attributes.Smoking','attributes.Take-out','attributes.Waiter Service']]
business_basics
```

```python
business_data=business_basics.rename(columns={'business_id':'business_id','categories':'categories','city':'city','full_address':'address','latitude':'lat','longitude':'long','name':'restaurant_name','neighborhoods':'neighborhood','review_count':'review_count','stars':'stars','state':'restaurant_state','attributes.Delivery':'delivery','attributes.Dogs Allowed':'dogs_allowed','attributes.Drive-Thru':'drive_thru','attributes.Happy Hour':'happy_hour','attributes.Noise Level':'noise_level','attributes.Outdoor Seating':'outdoor_seating','attributes.Price Range':'price_range','attributes.Takes Reservations':'take_reservations','attributes.Wi-Fi':'wi_fi','attributes.Smoking':'smoking','attributes.Take-out':'take_out','attributes.Waiter Service':'waiter_service'})
business_data
```

```python
yelp_data = business_data[['business_id', 'restaurant_state','city', 'restaurant_name','address','price_range','stars','neighborhood', 'dogs_allowed', 'drive_thru', 'happy_hour','delivery', 'noise_level',
       'outdoor_seating','take_reservations', 'wi_fi',
       'categories', 'lat', 'long',
        'review_count','smoking', 'take_out',
       'waiter_service']]
yelp_data
```

```python
# filter restaurants of US
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
usa=business1.loc[business['state'].isin(states)]
yelp_data.head()
```

```python
plt.figure(figsize=(11,6))
grouped = yelp_data.stars.value_counts().sort_index()
sns.barplot(grouped.index, grouped.values, palette=sns.color_palette("RdBu_r", len(grouped)))
plt.xlabel('Average Rating', labelpad=10, fontsize=14)
plt.ylabel('Count of restaurants', fontsize=14)
plt.title('Count of Restaurants against Ratings', fontsize=15)
plt.tick_params(labelsize=14)
for  i, v in enumerate(grouped):
    plt.text(i, v*1.02, str(v), horizontalalignment ='center',fontweight='bold', fontsize=14)
```

```python
plt.figure(figsize=(11,6))
grouped = yelp_data.groupby('city')['review_count'].sum().sort_values(ascending=False)[:10]
sns.barplot(grouped.index, grouped.values, palette=sns.color_palette("GnBu_r", len(grouped)) )
plt.xlabel('City', labelpad=10, fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Count of Reviews by City (Top 10)', fontsize=15)
plt.tick_params(labelsize=14)
plt.xticks(rotation=15)
for  i, v in enumerate(grouped):
    plt.text(i, v*1.02, str(v), horizontalalignment ='center',fontweight='bold', fontsize=14)
```

```python
plt.figure(figsize=(11,7))
grouped = yelp_data.groupby('city')['stars'].mean().sort_values(ascending = False)[:20]
sns.barplot(y=grouped.index, x= grouped.values, palette= sns.color_palette("RdBu_r", len(grouped)) )
plt.ylabel('City', fontsize=14)
plt.xlabel('Rating', fontsize=14)
plt.title('Average Rating by City', fontsize=15)
for i,v in enumerate(grouped):
    plt.text(v, i+0.15, str(v),fontweight='bold', fontsize=14)
```

```python
plt.figure(figsize=(11,7))
grouped = yelp_data.groupby('city')['price_range'].mean().sort_values(ascending = True)[:20]
sns.barplot(y=grouped.index, x= grouped.values, palette= sns.color_palette("RdBu_r", len(grouped)) )
plt.ylabel('City', fontsize=14)
plt.xlabel('Price Range', fontsize=14)
plt.title('Average price range by City', fontsize=15)
for i,v in enumerate(grouped):
    plt.text(v, i+0.15, str(v),fontweight='bold', fontsize=14)
```

```python
table = yelp_data.groupby(['city','price_range'])['stars'].mean().round().unstack()[:20]
plt.figure(figsize=(11,8))
sns.heatmap(table, cmap='YlGnBu', fmt='g',annot=True, linewidths=1)
plt.tick_params(labelsize=15)
```
